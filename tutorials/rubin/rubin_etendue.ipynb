# Calculating the Étendue of the Vera C. Rubin Observatory

Course: EDUCADO 2026
Topic: Survey telescope design and figures of merit

Étendue ($A\Omega$) quantifies a telescope's survey power. It is the product of the
effective collecting area $A_{\mathrm{eff}}$ and the solid angle $\Omega$ of the field of view:

$$A\Omega = A_{\mathrm{eff}} \times \Omega$$

This notebook walks through each step of the calculation to derive the Rubin Observatory canonical value of ≈ 319 m² deg².

Primary references used in this notebook:

**[LOP]** Ivezić, Ž. et al. 2019, *ApJ*, 873, 111 — *"LSST: From Science Drivers to Reference Design and Anticipated Data Products"*
arXiv:0805.2366 · DOI:10.3847/1538-4357/ab042c · short URL: `ls.st/lop`
→ Source for: étendue (319 m²·deg²), effective collecting area (35 m²), FOV diameter (3.5°)

**[RKN]** Rubin Observatory Key Numbers: https://rubinobservatory.org/for-scientists/rubin-101/key-numbers
→ Source for: étendue, effective collecting area, FOV

**[MIR]** Rubin Observatory Mirror page: https://www.lsst.org/about/tel-site/mirror
→ Source for: M1 outer diameter (8.4 m glass blank), M1 inner diameter, M2 diameter

**[CLV]** Claver, C. F. et al. — *"An Overview of the LSST Telescope and Site"*
https://www.lsst.org/sites/default/files/docs/Claver_086.18.pdf
→ Source for: individual mirror dimensions and optical design geometry

> **Note:** The individual mirror diameters (M1, M2, M3) are **not** in LOP Table 1.
> LOP Table 1 gives high-level system parameters only (effective aperture, collecting area, FOV, étendue).
> Mirror dimensions come from [MIR] and [CLV].

```python
import numpy as np
import pandas as pd

import holoviews as hv
from holoviews import opts
from bokeh.models import Div, HoverTool
hv.extension('bokeh')
```

## 1. Telescope Parameters

Rubin uses a unique three-mirror anastigmat design where the primary mirror (M1)
and the tertiary mirror (M3) are polished onto the same monolithic glass substrate.
This means M1 is an annulus and does not extend to the centre.

```python
# Mirror dimensions (metres)
# Sources: [MIR] Rubin Observatory Mirror page, [CLV] Claver et al.
D_M1_outer = 8.4      # M1 glass blank outer diameter [MIR]
D_M1_inner = 5.012    # M1 inner diameter / M3 outer diameter [CLV]
D_M2       = 3.42     # M2 (secondary) diameter [MIR, CLV]

# Field of view — from LOP Table 1 and [RKN]
fov_diameter_deg = 3.5

print(f"M1 outer diameter : {D_M1_outer} m   (source: [MIR, CLV])")
print(f"M1 inner diameter : {D_M1_inner} m  (source: [CLV])")
print(f"M2 diameter       : {D_M2} m   (source: [MIR, CLV])")
print(f"FoV diameter      : {fov_diameter_deg}°     (source: LOP Table 1, [RKN])")
```

```python
# Additional obscuration fraction (camera body, spider vanes, baffles)
additional_obscuration = 0.07  # ~7 %
```

## 2. Geometric Collecting Area of the M1 Annulus

Because M2 (3.42 m) is *smaller* than the M1 central hole (5.012 m), the secondary
does not create any extra obscuration; the hole in M1 already accounts for it.

$$A_{M1} = \frac{\pi}{4}\left(D_{\mathrm{outer}}^{2} - D_{\mathrm{inner}}^{2}\right)$$

```python
A_M1 = (np.pi / 4) * (D_M1_outer**2 - D_M1_inner**2)

print(f"M1 annulus area = (π/4) × ({D_M1_outer}² − {D_M1_inner}²)")
print(f"                = (π/4) × ({D_M1_outer**2:.2f} − {D_M1_inner**2:.3f})")
print(f"                = (π/4) × {D_M1_outer**2 - D_M1_inner**2:.3f}")
print(f"                = {A_M1:.2f} m²")
```

## 3. Effective Collecting Area

We apply a correction for the additional obscuration from the camera body,
spider vanes, and baffling (~7 %):

$$A_{\mathrm{eff}} = A_{M1} \times (1 - f_{\mathrm{obsc}})$$

```python
A_eff = A_M1 * (1 - additional_obscuration)

print(f"Additional obscuration : {additional_obscuration*100:.0f} %")
print(f"A_eff = {A_M1:.2f} × {1 - additional_obscuration:.2f} = {A_eff:.2f} m²")
```

## 4. Field-of-View Solid Angle

The focal plane subtends a 3.5° diameter circle on the sky.

### Flat-sky (small-angle) approximation

$$\Omega \approx \pi \left(\frac{\theta_{\mathrm{FoV}}}{2}\right)^{2} \quad [\mathrm{deg}^{2}]$$

### Exact spherical formula

$$\Omega_{\mathrm{exact}} = 2\pi\left(1 - \cos\frac{\theta_{\mathrm{FoV}}}{2}\right) \quad [\mathrm{sr}]$$

```python
# --- In square degrees (flat-sky) ---
half_angle_deg = fov_diameter_deg / 2
Omega_deg2 = np.pi * half_angle_deg**2

# --- Exact spherical ---
half_angle_rad = np.radians(half_angle_deg)
Omega_sr = 2 * np.pi * (1 - np.cos(half_angle_rad))
Omega_sr_to_deg2 = Omega_sr * (180 / np.pi)**2

print(f"Half-angle           : {half_angle_deg}° = {half_angle_rad:.6f} rad")
print(f"Ω (flat-sky approx)  : {Omega_deg2:.4f} deg²")
print(f"Ω (exact spherical)  : {Omega_sr_to_deg2:.4f} deg²")
print(f"\nDifference           : {abs(Omega_deg2 - Omega_sr_to_deg2):.4f} deg² "
      f"({abs(Omega_deg2 - Omega_sr_to_deg2)/Omega_deg2*100:.3f} %)")
print("\n→ At 3.5° the flat-sky approximation is excellent.")
```

## 5. Étendue Calculation

$$A\Omega = A_{\mathrm{eff}} \times \Omega$$

```python
etendue = A_eff * Omega_deg2
print(f"  A_eff  = {A_eff:.2f} m²")
print(f"  Ω      = {Omega_deg2:.4f} deg²")
print(f"  ────────────────────────────────")
print(f"  AΩ     = {etendue:.1f} m² deg²")
```

## 6. Sensitivity to Assumptions

The canonical 319 value depends on the assumed additional obscuration.
Let's see how the étendue varies.

```python
print(f"{'Obscuration (%)':>16}  {'A_eff (m²)':>11}  {'AΩ (m² deg²)':>13}")
print("-" * 46)

for obsc_pct in [0, 3, 5, 7, 10]:
    A = A_M1 * (1 - obsc_pct / 100)
    AO = A * Omega_deg2
    marker = "  ← canonical" if obsc_pct == 7 else ""
    print(f"{obsc_pct:>15.0f}%  {A:>11.2f}  {AO:>13.1f}{marker}")
```

```python
# Prepare data 
plot_df = df_tel.copy().reset_index(drop=True)
plot_df['x'] = range(len(plot_df))

CAT_COLORS = {
    SURVEY: '#4488bb',
    SPACE:  '#e07a2f',
    NARROW: '#9467bd',
    FUTURE: '#d62728'
}
RUBIN_COLOR = '#2ca02c'

# Per-bar fill: category colour; Rubin overridden to green
plot_df['Color'] = plot_df['Category'].map(CAT_COLORS)
plot_df.loc[plot_df['Telescope'].str.contains('Rubin'), 'Color'] = RUBIN_COLOR
plot_df['Label'] = plot_df.apply(lambda r: f"{r['Year']} · {r['Telescope']}", axis=1)
plot_df['etendue_val'] = plot_df['Étendue (m²·deg²)']
plot_df['a_eff_val']   = plot_df['A_eff (m²)']
plot_df['fov_val']     = plot_df['FOV (deg²)']

def format_etendue(row_data):
    value = row_data['Étendue (m²·deg²)']
    name  = str(row_data['Telescope']).strip()
    if abs(value) < 1e-9:
        return ""
    if "JWST" in name:
        return f"{value:.2f}"
    formatted = f"{value:.1f}"
    return formatted.rstrip('0').rstrip('.')

plot_df['text_label'] = plot_df.apply(format_etendue, axis=1)
plot_df['label_y']    = plot_df['Étendue (m²·deg²)'] * 1.3

Y_MIN, Y_MAX = 0.01, 1000
ALL_XTICKS   = [(i, lbl) for i, lbl in enumerate(plot_df['Label'])]

# Legend hook
def add_category_legend(plot, element):
    p = plot.state
    for leg in list(p.legend):
        leg.visible = False

    all_items = list(CAT_COLORS.items()) + [('Rubin Observatory', RUBIN_COLOR)]
    swatches = '&nbsp;&nbsp;&nbsp;'.join(
        f'<span style="color:{color}; font-size:15px;">&#9632;</span>'
        f'&nbsp;<span style="font-size:12px;">{cat}</span>'
        for cat, color in all_items
    )
    html = (
        '<div style="text-align:center; font-family:Arial,sans-serif; padding:4px 0;">'
        + swatches + '</div>'
    )
    p.add_layout(Div(text=html, sizing_mode='stretch_width'), 'above')

# Custom hover 
hover_tool = HoverTool(tooltips=[
    ('Telescope',         '@{Telescope}'),
    ('Category',          '@{Category}'),
    ('A_eff (m²)',         '@{a_eff_val}{0.2f}'),
    ('FOV (deg²)',         '@{fov_val}{0.4f}'),
    ('Étendue (m²·deg²)', '@{etendue_val}{0.2f}'),
])

# Chart bars
bars = hv.Bars(
    plot_df,
    kdims=['x'],
    vdims=['etendue_val', 'Color', 'Telescope', 'Category', 'a_eff_val', 'fov_val'],
).opts(opts.Bars(
    color=hv.dim('Color'),
    line_color='black',
    line_width=0.5,
    width=900,
    height=520,
    tools=[hover_tool, 'save'],
    toolbar='above',
    xlabel='',
    ylabel='Étendue (m²·deg²) [log scale]',
    title='Survey Telescope Étendue',
    fontsize={'title': 13, 'labels': 11, 'ticks': 9},
    show_grid=True,
    gridstyle={'grid_line_alpha': 0.3},
    xrotation=45,
    logy=True,
    ylim=(Y_MIN, Y_MAX),
    xticks=ALL_XTICKS,
    hooks=[add_category_legend],
))

# Value labels 

labels = hv.Labels(
    plot_df,
    kdims=['x', 'label_y'],
    vdims=['text_label'],
).opts(opts.Labels(
    text_font_size='9pt',
    text_align='center',
    text_color='black',
))

bars * labels
```

## Summary

Using Rubin's M1 annulus geometry ($D_{\mathrm{outer}} = 8.4$ m, $D_{\mathrm{inner}} = 5.012$ m)
with ~7 % additional obscuration and a 3.5° circular field of view:

$$\boxed{A\Omega \approx 319 \; \mathrm{m^{2}\,deg^{2}}}$$

This is roughly an **order of magnitude** larger than any previous survey telescope.
